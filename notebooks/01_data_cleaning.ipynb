### 🧹 Data Cleaning Steps

1. **Handling Missing Values in `CouponCode`**
   * **Action:** Replaced `NaN` values with `"No Coupon"`.
   * **Rationale:** A logical check proved that `TotalPrice == Quantity * UnitPrice` for these rows, confirming no discount was applied. Thus, the missing values simply represent orders made without a coupon code.

2. **Handling Duplicates in `CustomerID`**
   * **Action:** Retained all duplicate rows.
   * **Rationale:** Duplicate IDs are legitimate in this context, as they represent repeat purchases/multiple item orders by the same customer. Removing them would cause critical data loss.

3. **Value Consistency & Quality Check**
   * Verified all categorical and textual values; no anomalies or unintended tokens (e.g., `dirty_token`) were detected across the dataset.
